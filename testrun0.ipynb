```python
from ase.io import read, write

mol = read("start.xyz")

symbols = mol.get_chemical_symbols()

# remove H atoms attached to sulfur environment (simplified cleanup)
new_atoms = []
for atom in mol:
    if atom.symbol == "H":
        continue
    new_atoms.append(atom)

write("noH.xyz", new_atoms)
```

```python
from ase.io import read, write
import numpy as np

mol = read("start.xyz")

# indices of sulfur atoms (you already found them)
S1 = 8
S2 = 147

# get positions
pos = mol.get_positions()

# force S–S distance ~2.05 Å (disulfide bond length)
distance = 2.05

vec = pos[S2] - pos[S1]
vec = vec / np.linalg.norm(vec)

pos[S2] = pos[S1] + distance * vec

mol.set_positions(pos)

write("cyclic.xyz", mol)
```
